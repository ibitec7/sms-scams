```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lr = [0.9220338983050848, 0.912751677852349, 0.9315068493150684]
rf = [0.8796992481203008, 0.785234899328859, 1]
nb = [0.8055555555555556, 0.9731543624161074, 0.6872037914691943]
svc = [0.9456992380239887, 0.9348210290827741, 0.9922916666666666]
knn = [0.9319202758502783, 0.8512639821029082, 1]
```

```python
results = [lr,rf,nb,svc,knn]

f1 = []
recall = []
precision = []

for model in results:
    f1.append(model[0])
    recall.append(model[1])
    precision.append(model[2])

idx = ["Logistic Regression", "Random Forest", "Naive Bayes", "SVC", "KNN"]

df_results = pd.DataFrame({"F1 score": f1, "Recall score": recall, "Precision score": precision}, index=idx)
```

```python
sns.heatmap(data=df_results, annot=True, fmt=".4f")
plt.title("Performance Metrics for Models")
```
